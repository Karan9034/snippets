```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_squared_error
```

```python
X, y = fetch_california_housing(return_X_y=True)
arr = np.transpose(np.array([np.ones(X.shape[0], dtype = int)]))
X = np.append(arr, X, axis=1)
y = np.reshape(y, (y.shape[0], 1))
print(X)
print(y)
```

```python
def normal_equation(X, y):
    X_trans = X.T
    inv = np.linalg.pinv(X_trans.dot(X))
    theta = inv.dot(X_trans).dot(y)
    return theta
```

```python
def fit(X, y):
    theta = normal_equation(X, y)
    return theta
```

```python
def predict(data):
    pred = np.dot(data, theta)
    return pred
```

```python
theta = fit(X, y)
```
